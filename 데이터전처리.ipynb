```python
# 이상치 데이터 검색 함수 정의
def find_outliers(df) :
    outlier_list = []

    for feature in df.columns[:-1] :
        q1 = df[feature].quantile(0.25)                             #1Q값 계산
        q3 = df[feature].quantile(0.75)                             #3Q값 계산

        iqr = q3 - q1                                               #IQR값 계산

        low_bound = q1 - 1.5 * iqr                                  #아랫경계값 계산
        up_bound = q3 + 1.5 * iqr                                   #윗경계값 계산

        outlier = df[(df[feature] < low_bound) | (df[feature] > up_bound)]    # 이상치 검색
        # 이상치 데이터의 인덱스 번호 저장
        outlier_list.extend(outlier.index)
        print(f'feature : {feature}, Low Bound : {low_bound}, Up Bound : {up_bound}')
    #인덱스 리스트에서 중복값 제거
    outlier_list = list(set(outlier_list))
    
    return outlier_list    
```

```python
iris_t2 = iris_df[iris_df['target']==2]
outl_idx = find_outliers(iris_t2)
print(outl_idx)
#print("target=2\n", iris_df.loc[outl_idx[:][0]])
```

```python
#타겟별 이상치 검색
iris_t0 = iris_df[iris_df['target']==0]
outl_idx = find_outliers(iris_t0)
print("target=0\n", iris_df.loc[outl_idx])

iris_t1 = iris_df[iris_df['target']==1]
outl_idx = find_outliers(iris_t1)
print("target=1\n", iris_df.loc[outl_idx])

iris_t2 = iris_df[iris_df['target']==2]
outl_idx = find_outliers(iris_t2)
print("target=2\n", iris_df.loc[outl_idx])
```

```python
#데이터 병합
df1 = pd.DataFrame({
    'id' : [1,2,3,4],
    'name' : ['Kim','Lee','Park','Yoo'],
    'age' : [25,30,35,40]
})

df2 = pd.DataFrame({
    'id' : [3,4,5,6],
    'city' : ['Daejeon','Seoul','Busan','Daegu'],
    'salary' : ['200','250','220','210']
})
```

```python
#inner join
merged_df_inner = pd.merge(df1, df2, on='id', how='inner')
merged_df_inner
```

```python
#left join
merged_df_left = pd.merge(df1, df2, on='id', how='left')
merged_df_left
```

```python
#right join
merged_df_right = pd.merge(df1, df2, on='id', how='right')
merged_df_right
```

```python
#outer join
merged_df_outer = pd.merge(df1, df2, on='id', how='outer')
merged_df_outer
```
